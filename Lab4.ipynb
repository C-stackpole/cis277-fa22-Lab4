## NFL Data

1. Import the NFL week 3, 4, and 5 data and add a week column to dataset and set the value to 3, 4, or 5 for all rows from that week.

```python
week_3 = pd.read_csv("nfloffenseweek3.csv")
week_3["week"] = 3
week_3
```

```python
week_4 = pd.read_csv("nfloffenseweek4.csv")
week_4["week"] = 4
week_4
```

```python
week_5 = pd.read_csv("nfloffenseweek5.csv")
week_5["week"] = 5
week_5
```

2. Combine the weekly dataframes by adding the dataframes together row-wise.

```python
nfl_stats = pd.concat([week_3, week_4, week_5])
nfl_stats
```

3. Make up one intresting question about this data set and answer it.

What teams were the most efficient on comparison of points per play?

```python
nfl_stats["points_per_play"]  = nfl_stats["points_scored"]/nfl_stats["offensive_plays"]
ppp_df = nfl_stats.drop(["games","total_yards", 'yards_per_play', 'turnovers_lost', 'fumbles_lost',
       '1st_downs', 'passes_completed', 'passes_attempted', 'passing_yards',
       'passing_touchdowns', 'passing_interceptions',
       'net_yards_per_pass_attempt', 'passing_1st_downs', 'rushing_attempts',
       'rushing_yards', 'rushing_touchdowns', 'rushing_yards_per_attempt',
       'rushing_1st_downs', 'penalties', 'penalty_yards', '1st_down_penalties',
       'percentage_scoring_drives', 'percentage_turnover_drives',
       'expected_points'], axis=1)
ppp_df.sort_values(["week","points_per_play"], ascending=False)
```

We see that teams that finished with higher ranks had more points per play, but we do see the case where the Dolphins had been ranked 8th offensivesly but were 2nd most effiecnt in week 3. I feel like teams should be ranked more on efficiency rather than only who scored the most points, because yoi could score a bunch of points yet be very unefficient.

## World Data

```python
height_weight = pd.read_csv("height_weight_data.csv")
iq = pd.read_csv("iq.csv")
life_expec = pd.read_csv("life_expectancy.csv")
pop_density = pd.read_csv("population_density.csv")
quality_life = pd.read_csv("quality_of_life.csv")
```

1. Join the height and weight data to the iq data.

```python
height_weight.merge(iq, how="left", on="country")
```

2. Join poulation density to quality of life.

```python
pd_ql_df = quality_life.merge(pop_density,how="left", on="country")
pd_ql_df
```

3. Join life expectncy to quality of life.

```python
quality_life.merge(life_expec,how="left", on="country")
```

4. Join iq, life expectncy, quality of life, and population density into one data frame

```python
le_iq_df = iq.merge(life_expec, how="right", on="country")
pd_ql_df.merge(le_iq_df, how="left", on="country")
```

5. Make up one intresting question about this data set and answer it.

It appears to be the case with Hong Kong having on of the highest iq's and also having the longest life expectanies for both male and female. But also we see just in general that countries that have higher iq's normally have longer life expecancies.
